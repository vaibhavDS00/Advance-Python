```python
class computer:
    def __init__(self, cpu, ram, rom):
        self.cpu = cpu
        self.ram = ram
        self.rom = rom
    def config(self):
        print("Config is :",self.cpu,self.ram,self.rom)

comp1=computer("intel -i7",'16 GB', '1 TB') #Object of computer type designes
comp2=computer('Exynos','32 GB','5 TB')

computer.config(comp1)
computer.config(comp2)
```
