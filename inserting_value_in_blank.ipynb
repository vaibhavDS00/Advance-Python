```python
data = [5, 7, 8, 9, None, 1]

print("Original Data:", data)

total = 0
count = 0
for value in data:
    if value is not None:
        total += value
        count += 1

mean = total / count


for i in range(len(data)):
    if data[i] is None: 
        data[i] = mean

print(data)
```

```python
data = [5, 7, 8, 9, None, 1]

print("Original Data:", data)

clean_data = [x for x in data if x is not None]
mean = sum(clean_data) / len(clean_data)

for i in range(len(data)):
    if data[i] is None:
        data[i] = mean

print(data)
```
