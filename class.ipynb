```python
num = int(input("Enter a number: ")) 
fixed_num = num                   
reverse = 0                  
while num > 0:
    remainder = num % 10
    reverse = reverse * 10 + remainder
    num //= 10
print("rev:", reverse)
if fixed_num == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```
