```python
class student:
    school = "AAFT"
    def __init__(self,m1,m2,m3):
        self.m1=m1
        self.m2=m2
        self.m3=m3

    def avg(self):
        return (self.m1+self.m2+self.m3)/3
    
    def get_m1(self):
        return self.m1
    
    def set_m1(self,value):
        self.m1 = value

    @classmethod
    def get_school(cls):
        return cls.school
    
    @staticmethod
    def info():
        print('This is student class... being demonstrated the class instructer')


s1=student(35,74,65)
s2=student(25,54,75)
print(s1.avg())
print(s2.avg())
print(student.get_school())
s1.info()
```

```python
class A: # Parent class/ Super class
    def feat1(self):
        print("Feature 1 working")
    def feat2(self):
        print("Feature 2 working")
    def feat3(self):
        print("Feature 3 working")
class B(A): # Child class/ Sub class
    def feat4(self):
        print("Feature 4 working")
    def feat5(self):
        print("Feature 5 working")

class C(B): # Sub class of B and Grand child of A
    def feat6(self):
        print("Feature 6 working")
    def feat7(self):
        print("Feature 7 working")

a1=A()
b1=B()
c1=C()

c1.feat1() # Inherited from class A
c1.feat2() # Inherited from class A
c1.feat3() # Inherited from class A
c1.feat4() # Inherited from class B
c1.feat5() # Inherited from class B
c1.feat6() # Own feature of class C
c1.feat7() # Own feature of class C
```

```python
# Single level inheritance

class A:   #Parent Class/Super Class
    def feat1(self):
        print("Feature 1 working")
    def feat2(self):
        print("Feature 2 working")
    def feat3(self):
        print("Feature 3 working")

class B(A):  #Child class / Sub calss
    def feat4(self):
        print("Feature 4 working")
    def feat5(self):
        print("Feature 5 working")

a1=A()
b1=B()

a1.feat1()
b1.feat1()
```

```python
# Multi Level Inheritance
class A:   #Parent Class/Super Class
    def feat1(self):
        print("Feature 1 working")
    def feat2(self):
        print("Feature 2 working")
    def feat3(self):
        print("Feature 3 working")

class B(A):  #Child class / Sub calss
    def feat4(self):
        print("Feature 4 working")
    def feat5(self):
        print("Feature 5 working")

class C(B):  #Child class / Sub calss
    def feat6(self):
        print("Feature 4 working")
    def feat7(self):
        print("Feature 5 working")

a1=A()
b1=B()
c1=C()

a1.feat3()
b1.feat2()
c1.feat1()
```

```python
# Multiple Inheritance

class A:   #Parent Class/Super Class
    def feat1(self):
        print("Feature 1 working")
    def feat2(self):
        print("Feature 2 working")
    def feat3(self):
        print("Feature 3 working")

class B:  #Child class / Sub calss
    def feat4(self):
        print("Feature 4 working")
    def feat5(self):
        print("Feature 5 working")

class C(B,A):  #Child class / Sub calss
    def feat6(self):
        print("Feature 4 working")
    def feat7(self):
        print("Feature 5 working")

a1=A()
b1=B()
c1=C()


c1.feat5()
```

```python
class A: #Parent / Super class
    def __init__(self):
        print("in init block A")
    def feat1(self):
        print("Feat1 working")

class B(A): #Child / Sub class
    def __init__(self):
        super().__init__() #Super.init method
        print("in init block B")
    def feat2(self):
        print("Feat2 working")

a=B()  # This state will at first try to search intin in child block itself, If not found then moves inside parent class

a.feat1()
```
