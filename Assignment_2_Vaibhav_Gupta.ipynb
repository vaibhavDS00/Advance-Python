```python
# 1. Implement computing factorial of a number using for loop and any 5 pattern related problems in python language.
num = 5  # Example number
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print(f"Factorial of {num} is {factorial}")

# Pattern 1: Right Angle Triangle
for i in range(1, n + 1):
    print('*' * i)
```

```python
# Pattern 2: Inverted Right Angle Triangle
for i in range(n, 0, -1):
    print('*' * i)
```

```python
# Pattern 3: Pyramid 
for i in range(n):
    print(' '*(n-i-1) + '*'*(2*i+1))
```

```python
# Pattern 4: Diamond
for i in range(n):
    print(' '*(n-i-1) + '*'*(2*i+1))
```

```python
# Pattern 5: Square
n = 5
for i in range(n):
    print('* ' * n)
```

```python
# Implement a Python program demonstrating Object-Oriented Programming concepts including classes.
# inheritance (single/multiple), encapsulation, polymorphism, abstraction, and advanced features like iterators, generators, decorators, and context managers.
# Use getter/setter methods, inbuilt functions, and recursive functions where applicable.

# Single level inheritance

class AAFT:   #Parent Class/Super Class
    def f1(self):
        print("The Student")
    def f2(self):
        print("Name is Vaibhav Gupta")

class doing(AAFT):  #Child class / Sub calss
    def f3(self):
        print("He is doing")
    def f4(self):
        print("Diploma in Data Science")

a1=AAFT()
b1=doing()

b1.f1()
b1.f2()
b1.f3()
b1.f4()
```

```python
# Multiple Inheritance

class student:   #Parent Class/Super Class
    def f1(self):
        print("My name is")
    def f2(self):
        print("Vaibhav Gupta")
    def f3(self):
        print("from delhi")

class college:  #Child class / Sub calss
    def f4(self):
        print("I've studied in")
    def f5(self):
        print("IMS Ghaziabad university courses campus")

class course(student,college):  #Child class / Sub calss
    def f6(self):
        print("I am doing")
    def f7(self):
        print("Bachelor's of Computer Applications")

a1=student()
b1=college()
c1=course()


c1.f1()
c1.f2()
c1.f3()
c1.f4()
c1.f5()
c1.f6()
c1.f7()
```

```python
# Plymorphism
class AAFT:
    def course(self):
        print("AAFT offers various courses.")
    def language(self):
        print("AAFT primarily uses English as the medium of instruction.")
class IIT:
    def course(self):
        print("IIT offers engineering and technology courses.")
    def language(self):
        print("IIT primarily uses English as the medium of instruction.")

obj_AAFT = AAFT()
obj_IIT = IIT()

for courses in (obj_AAFT, obj_IIT):
    courses.course()
    courses.language()
```
